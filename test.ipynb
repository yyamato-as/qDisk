```python
from qdisk.plot import Map
import matplotlib.pyplot as plt

imagename = "../eDisk/firstlook_L1489IRS/VADP/L1489IRS_SBLB_C18O_robust_1.0.image_v0.fits"
vsys = 7.22

%matplotlib widget
%load_ext autoreload
%autoreload 2
```

```python
fig, ax = plt.subplots(figsize=(5,4))

rotmap = Map(imagename, ax=ax, data_scaling_factor=1e-3, xlim=(-5, 5), ylim=(-5, 5))

rotmap.mask(rmax=5.0, rmin=2.0, vmin=4, vmax=11, combine="or")
rotmap.plot_colormap(cmap="RdBu_r", vmin=vsys-5, vmax=vsys+5)
rotmap.add_beam(color="black")
rotmap.add_scalebar(color="black", scale=50/140, text="50 au")
rotmap.add_colorbar(label="km s$^{-1}$", rotation=270)
rotmap.set_ticker(minor=True, majornticks=5, minornticks=5)
rotmap.set_labels(xlabel="$\Delta$R.A. [arcsec]", ylabel="$\Delta$Dec. [arcsec]")
```
